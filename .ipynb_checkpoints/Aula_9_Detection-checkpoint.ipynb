## Introdução

Nosso objetvo é apresentar no dominio de detecção de objetos o desenvolvimento das redes R-CNN, Fast-RCNN e Faster-RCNN.


# Redes R-CNNs

As redes R-CNN (Region-based Convolutional Neural Networks) surgiram como um marco em visão computacional, oferecendo uma nova perspectiva para resolver os desafios de **Detecção de Objetos** ao combinar o poder das redes neurais convolucionais (CNNs) com métodos de proposição de regiões.

## 1. R-CNN (Region-based Convolutional Neural Network)

A R-CNN foi uma das primeiras arquiteturas a utilizar redes neurais convolucionais (CNNs) para a tarefa de detecção de objetos, combinando o poder da extração de features profundas com métodos clássicos de proposição de regiões. A arquitetura da R-CNN pode ser dividida em quatro componentes principais:

1.  **Geração de Propostas de Região:** O primeiro passo da R-CNN é gerar um conjunto de regiões de interesse (Region Proposals) na imagem de entrada que provavelmente contêm objetos. Métodos como o *Selective Search* são comumente usados para essa finalidade. O Selective Search analisa a imagem e agrupa pixels com base em características de cor, textura e forma, gerando um conjunto de cerca de 2000 propostas de região por imagem. A ideia é que, se houver um objeto na imagem, ele estará contido em uma ou mais dessas propostas.

2.  **Extração de Features com CNN:** Cada proposta de região gerada na etapa anterior é redimensionada para um tamanho fixo (geralmente 224x224 pixels, o tamanho de entrada de CNNs populares como AlexNet ou VGG). Essa região redimensionada é então passada por uma rede neural convolucional pré-treinada (em grandes datasets como ImageNet) para extrair um vetor de features de alta dimensão. A CNN atua como um extrator de features poderoso, aprendendo representações hierárquicas da imagem.

3.  **Classificação com SVMs:** O vetor de features extraído pela CNN para cada proposta de região é então alimentado em um conjunto de classificadores *Support Vector Machine (SVM)*. Cada SVM é treinado para classificar se a proposta de região contém um objeto de uma classe específica (e.g., carro, pessoa, gato) ou se é background. O uso de SVMs, em vez de uma camada softmax no final da CNN, foi justificado na época pela melhora no desempenho de classificação, especialmente para classes com menos exemplos de treinamento.

4.  **Refinamento com Bounding Box Regressor:** Para cada proposta de região classificada como contendo um objeto por um dos SVMs, um regressor linear de caixa delimitadora (Bounding Box Regressor) é utilizado. Este regressor é treinado para ajustar as coordenadas da caixa delimitadora proposta para que ela se ajuste mais precisamente à localização real do objeto na imagem. Isso ajuda a corrigir pequenas imprecisões nas propostas de região iniciais.

### Vantagens e Desvantagens da R-CNN

**Vantagens:**

*   **Uso de CNNs:** A principal vantagem da R-CNN foi a incorporação das CNNs para extração de features, o que resultou em um salto significativo na precisão da detecção de objetos em comparação com métodos baseados em features manuais. As CNNs são capazes de aprender representações robustas e discriminativas diretamente dos dados.

**Desvantagens:**

*   **Lenta:** O principal gargalo da R-CNN é a sua lentidão. A CNN precisa ser executada separadamente para cada uma das cerca de 2000 propostas de região por imagem. Isso torna o processo de detecção muito demorado, levando vários segundos por imagem, o que inviabiliza aplicações em tempo real.
*   **Treinamento Complexo:** O treinamento da R-CNN envolve múltiplas etapas separadas: ajuste fino da CNN, treinamento dos classificadores SVMs e treinamento dos bounding box regressors. Cada uma dessas etapas requer dados e configurações específicos, tornando o processo de treinamento complexo e difícil de otimizar de ponta a ponta.
*   **Armazenamento de Features:** Os vetores de features extraídos pela CNN para todas as propostas de região precisam ser armazenados em disco para treinar os SVMs e os regressors, o que demanda uma grande quantidade de espaço de armazenamento.

## 2. Fast R-CNN

O Fast R-CNN foi proposto como uma evolução do R-CNN, focado principalmente em resolver os problemas de velocidade e complexidade de treinamento da arquitetura original. Embora ainda utilize a ideia de propostas de região, o Fast R-CNN introduz otimizações cruciais que o tornam significativamente mais rápido e eficiente.

As principais melhorias introduzidas no Fast R-CNN são:

1.  **Execução da CNN uma Única Vez por Imagem:** Diferentemente do R-CNN, onde a CNN era executada separadamente para cada uma das milhares de propostas de região, o Fast R-CNN executa a rede convolucional **apenas uma vez** para a imagem de entrada completa. Isso gera um mapa de features (feature map) para a imagem inteira.
2.  **RoI Pooling (Region of Interest Pooling):** Esta é uma das inovações chave do Fast R-CNN. Após obter o mapa de features da imagem inteira, as propostas de região originais (que estavam na imagem de entrada) são projetadas de volta no mapa de features. A camada RoI Pooling então extrai um pequeno mapa de features (de tamanho fixo, por exemplo, 7x7) para cada proposta de região a partir do mapa de features global. Isso é feito dividindo a região projetada em uma grade de tamanho fixo e aplicando max-pooling em cada célula da grade. O resultado é um vetor de features de tamanho fixo para cada proposta de região, independentemente do seu tamanho original na imagem.
3.  **Treinamento Multi-Tarefa:** O Fast R-CNN treina a rede de ponta a ponta com uma única função de perda multi-tarefa. Esta função de perda combina duas tarefas principais:
    *   **Classificação:** Utiliza uma camada softmax para prever a classe de cada proposta de região (incluindo uma classe "background").
    *   **Regressão da Caixa Delimitadora:** Utiliza um regressor linear para refinar as coordenadas da caixa delimitadora para cada proposta de região classificada como objeto.
    A perda total é a soma da perda de classificação e da perda de regressão da caixa. Isso permite que a rede aprenda a extrair features e a realizar a classificação e a regressão simultaneamente, simplificando o treinamento.

### Diferenças e Vantagens do Fast R-CNN

As diferenças entre R-CNN e Fast R-CNN resultam em vantagens significativas:

*   **Velocidade:** A execução da CNN uma única vez e o uso eficiente do RoI Pooling tornam o Fast R-CNN muito mais rápido que o R-CNN, tanto no treinamento quanto na inferência. A inferência pode ser até 10 vezes mais rápida que a R-CNN.
*   **Treinamento Simplificado:** O treinamento multi-tarefa permite treinar a rede inteira em uma única etapa, eliminando a necessidade de treinar SVMs e regressors separadamente e armazenar features em disco.
*   **Melhor Precisão:** O treinamento de ponta a ponta permite que a rede otimize todos os seus componentes conjuntamente para a tarefa final de detecção, levando, em muitos casos, a uma melhor precisão.


## 3. Faster R-CNN

O Faster R-CNN representa um avanço significativo na família R-CNN ao integrar a etapa de proposição de regiões **dentro da própria rede neural**, eliminando a necessidade de algoritmos externos (como o Selective Search usado no R-CNN e Fast R-CNN). A principal inovação do Faster R-CNN é a introdução da **Region Proposal Network (RPN)**.

### Region Proposal Network (RPN)

A RPN é uma rede convolucional totalmente treinável que, simultaneamente, prevê **propostas de região** e seus respectivos **scores de "objeto"** em uma imagem. Ela opera diretamente sobre o mapa de features compartilhado gerado pela rede convolucional base (a mesma CNN usada para extrair features para detecção).

Como a RPN funciona:

1.  **Camadas Convolucionais Compartilhadas:** A imagem de entrada é primeiramente processada por uma série de camadas convolucionais que são compartilhadas tanto pela RPN quanto pela rede de detecção subsequente (essencialmente a parte do Fast R-CNN). Isso gera um mapa de features da imagem completa.
2.  **Pequena Rede Convolucional de Saída:** Sobre o mapa de features compartilhado, uma pequena rede convolucional desliza espacialmente. Para cada posição deslizante no mapa de features, essa rede gera duas saídas:
    *   **Camada de Classificação (cls layer):** Produz scores para k propostas de região simultaneamente em cada localização. Para cada proposta, ele estima a probabilidade de ser um "objeto" (foreground) ou "não-objeto" (background).
    *   **Camada de Regressão (reg layer):** Codifica as coordenadas das caixas delimitadoras para as k propostas em cada localização.
3.  **Anchor Boxes:** A chave para a RPN gerar múltiplas propostas em cada localização é o conceito de *anchor boxes*. Em cada posição espacial do mapa de features, a RPN considera k caixas de referência pré-definidas com diferentes tamanhos e proporções (aspect ratios). Essas são as "anchor boxes". A RPN prediz se cada uma dessas anchor boxes contém um objeto e, se sim, prediz os ajustes (deltas) necessários para as coordenadas da caixa para melhor se ajustar ao objeto real. Por exemplo, k=9 anchor boxes podem ser usadas em cada localização, consistindo de 3 tamanhos (e.g., 128, 256, 512 pixels de área) e 3 proporções (e.g., 1:1, 1:2, 2:1). Para um mapa de features de tamanho WxH, isso resulta em um total de W * H * k propostas de região.

### Integração da RPN na Rede Neural

A RPN é treinada de ponta a ponta juntamente com a rede de detecção. O fluxo geral do Faster R-CNN é:

1.  A imagem de entrada passa pelas **camadas convolucionais compartilhadas** para produzir um mapa de features.
2.  O **RPN** opera sobre este mapa de features para gerar um conjunto de **propostas de região** com seus scores de objeto.
3.  As propostas de região geradas pela RPN são então usadas pela camada **RoI Pooling** (ou RoI Align, em versões posteriores) para extrair vetores de features de tamanho fixo do mapa de features compartilhado para cada proposta.
4.  Esses vetores de features são alimentados nas **camadas totalmente conectadas** e, em seguida, nas camadas de **classificação** (para determinar a classe específica do objeto) e **regressão da caixa delimitadora** (para refinar as coordenadas da caixa).

Todo o pipeline (camadas compartilhadas, RPN, RoI Pooling, classificação e regressão final) pode ser treinado como uma única rede, embora o treinamento seja tipicamente realizado em múltiplas etapas para otimizar a RPN e a rede de detecção de forma alternada.

### Diferenças e Vantagens em Relação aos Modelos Anteriores

A principal diferença do Faster R-CNN em relação ao R-CNN e Fast R-CNN é a substituição do algoritmo externo de proposição de região pela RPN, que é parte integrante da rede neural.

As vantagens do Faster R-CNN são significativas:

*   **Velocidade:** A RPN é muito mais rápida na geração de propostas de região do que métodos externos como Selective Search. Como a RPN e a rede de detecção compartilham as camadas convolucionais, a maior parte da computação (a extração de features) é realizada apenas uma vez. Isso torna o Faster R-CNN dramaticamente mais rápido que o R-CNN e também mais rápido que o Fast R-CNN (que ainda dependia de um método de proposição externo lento), permitindo detecção em tempo quase real.
*   **Treinamento End-to-End:** Embora o treinamento possa envolver múltiplos passos alternados, o modelo geral é treinado de ponta a ponta, otimizando a geração de propostas e a detecção de objetos conjuntamente. Isso leva a um melhor desempenho.
*   **Precisão:** A integração da RPN no pipeline de treinamento permite que as propostas de região geradas sejam otimizadas para a tarefa de detecção de objetos, o que geralmente resulta em maior precisão em comparação com o uso de propostas agnósticas a objetos.


## Comparativo entre R-CNN, Fast R-CNN e Faster R-CNN

Para consolidar o entendimento das evoluções e diferenças entre as arquiteturas R-CNN, Fast R-CNN e Faster R-CNN.

| Característica                 | R-CNN                                   | Fast R-CNN                                    | Faster R-CNN                                      |
| :----------------------------- | :-------------------------------------- | :-------------------------------------------- | :------------------------------------------------ |
| **Componente de Proposição de Região** | Algoritmo externo (ex: Selective Search) | Algoritmo externo (ex: Selective Search)      | Rede Neural Integrada (Region Proposal Network - RPN) |
| **Execução da CNN Base**       | Uma vez por *proposta de região*        | Uma vez por *imagem inteira*                  | Uma vez por *imagem inteira* (compartilhada com RPN) |
| **Mecanismo de Pooling**       | Redimensionamento + CNN separada        | RoI Pooling                                   | RoI Pooling (ou RoI Align)                        |
| **Treinamento**                | Multistep separado (CNN, SVMs, Regressor) | Multi-tarefa end-to-end                       | Multi-tarefa end-to-end  |
| **Velocidade (Inferência)**    | Lento (segundos por imagem)             | Rápido (frações de segundo por imagem)        | Muito rápido (quase tempo real)                   |
| **Precisão**                   | Melhor que métodos tradicionais         | Melhor que R-CNN                              | Melhor que Fast R-CNN (geralmente)                |
| **Geração de Propostas**       | Lenta, independente da rede de detecção | Lenta, independente da rede de detecção       | Rápida, parte integrante e otimizada para a detecção |

**Resumo das Principais Conclusões do Comparativo:**

A evolução da família R-CNN demonstra uma clara progressão em direção à eficiência e integração. O **R-CNN** original validou o uso de CNNs para extração de features de objetos, mas sofreu com sua lentidão devido à execução repetida da CNN e ao pipeline de treinamento complexo. O **Fast R-CNN** resolveu a lentidão da extração de features compartilhando a computação da CNN e introduzindo o RoI Pooling, além de simplificar o treinamento com uma perda multi-tarefa. Finalmente, o **Faster R-CNN** deu o passo crucial de substituir o gargalo da proposição de regiões por um componente neural (RPN) treinado de ponta a ponta com a rede de detecção, alcançando velocidades próximas ao tempo real e estabelecendo um padrão para futuros detectores de objetos. Cada geração construiu sobre as inovações da anterior para criar sistemas de detecção de objetos mais rápidos, precisos e eficientes.

## Resumo

### Principais Conclusões da Análise

* A arquitetura **R-CNN** foi um passo fundamental, combinando *features* extraídas por CNN com propostas de regiões externas. Porém, era lenta, pois a CNN era executada repetidamente para cada região.
* A **Fast R-CNN** aumentou a velocidade ao executar a CNN apenas uma vez por imagem e usar **RoI Pooling**, além de simplificar o treinamento com uma **função de perda multi-tarefa**.
* A **Faster R-CNN** atingiu quase tempo real ao integrar a geração de propostas de regiões dentro da própria rede, através da **Region Proposal Network (RPN)**, eliminando o gargalo externo.
* Cada nova versão melhorou a **velocidade**, **eficiência** e geralmente a **precisão** em relação à anterior.

### Insights

* A evolução mostra uma tendência clara de **integração e otimização dos componentes** dentro de uma única rede treinável de ponta a ponta, resultando em melhor desempenho e velocidade na detecção de objetos.

```python
# Faster R-CNN com Transfer Learning
import torch
import torchvision
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Carregar modelo pré-treinado
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
model.eval()  # modo inferência

# Carregar imagem da URL
url = "https://cdn.pixabay.com/photo/2018/08/01/17/15/city-3577655_640.jpg"  # exemplo
response = requests.get(url)
img = Image.open(BytesIO(response.content)).convert("RGB")

# Transformar imagem em tensor
from torchvision import transforms
transform = transforms.Compose([
    transforms.ToTensor(),
])
img_tensor = transform(img)

# Inferência
with torch.no_grad():
    outputs = model([img_tensor])  # retorna uma lista de dicionários

# Mostrar resultados
output = outputs[0]
boxes = output['boxes']
labels = output['labels']
scores = output['scores']

# Filtrar resultados com confiança > 0.5
threshold = 0.5
boxes = boxes[scores > threshold]
labels = labels[scores > threshold]
scores = scores[scores > threshold]

# Mapear IDs para nomes COCO
COCO_INSTANCE_CATEGORY_NAMES = [
    '__background__', 'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
    'train', 'truck', 'boat', 'traffic light', 'fire hydrant', 'stop sign',
    'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow',
    'elephant', 'bear', 'zebra', 'giraffe', 'backpack', 'umbrella', 'handbag',
    'tie', 'suitcase', 'frisbee', 'skis', 'snowboard', 'sports ball', 'kite',
    'baseball bat', 'baseball glove', 'skateboard', 'surfboard', 'tennis racket',
    'bottle', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl', 'banana',
    'apple', 'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza',
    'donut', 'cake', 'chair', 'couch', 'potted plant', 'bed', 'dining table',
    'toilet', 'tv', 'laptop', 'mouse', 'remote', 'keyboard', 'cell phone',
    'microwave', 'oven', 'toaster', 'sink', 'refrigerator', 'book', 'clock',
    'vase', 'scissors', 'teddy bear', 'hair drier', 'toothbrush'
]

# Plotar imagem com caixas
fig, ax = plt.subplots(1, figsize=(12,9))
ax.imshow(img)

for box, label, score in zip(boxes, labels, scores):
    x1, y1, x2, y2 = box
    rect = patches.Rectangle(
        (x1, y1), x2-x1, y2-y1,
        linewidth=2, edgecolor='r', facecolor='none'
    )
    ax.add_patch(rect)
    ax.text(x1, y1, f"{COCO_INSTANCE_CATEGORY_NAMES[label]}: {score:.2f}",
            color='white', fontsize=12, backgroundcolor="red")

plt.axis('off')
plt.show()
```
